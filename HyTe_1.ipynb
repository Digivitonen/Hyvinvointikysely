```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
dat = pd.read_csv("https://raw.githubusercontent.com/rerkki/metropolia/master/esimerkkikysely2.csv", encoding='latin-1')

dat[0:1] #tulostetaan otsikko ja ensimmäinen rivi tarkistuksen vuoksi
```

```
 dat.rename(columns={'Tanaan herasin taynna energiaa': 'K1', 
                      'Minulla on suunnitelmia ensi viikonlopulle': 'K2',
                      'Olen liikkunut tanaan riittavasti': 'K3',
                      'Aion olla hyvassa kunnossa ensi kesana': 'K4',
                      'Nukuin viime yon surkeasti': 'K5',
                      'Haluaisin vaihtaa alaa': 'K6'})

dat[0:5] #tulostetaan otsikko ja ensimmäinen rivi tarkistuksen vuoksi
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(12,8)})

ax = sns.barplot(x=dat.kayttaja, y=dat.K1, data=dat)
```

```
sns.set(style="whitegrid", rc={'figure.figsize':(24,8)})

fig, ax =plt.subplots(1,2)
sns.lineplot(x=dat.Aikaleima, y=dat.K1, data=dat, ax = ax[0])
sns.lineplot(x= dat.index, y=dat.K1, data=dat, ax = ax[1])

fig.show()

```

```
dat2 = dat[['K1', 'K2', 'K3', 'K4', 'K5', 'K6']]

corr = dat2.corr()
mask = np.triu(np.ones_like(corr, dtype=np.bool))
f, ax = plt.subplots(figsize=(11, 9))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0, square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

```
sns.pairplot(dat2, corner = True)
```
